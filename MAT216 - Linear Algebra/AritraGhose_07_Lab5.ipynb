A set of vectors V = { $ v_{1} $ , $ v_{2} $, . . . . . . . , $ v_{n} $} are mutually orthogonal if every pair of
vectors is orthogonal. i.e. $ v_{i} $ . $ v_{j} $= 0, for all i â‰  j.

# Task 1
Check whether the set of vectors is orthogonal or not.

$$v_{1} = \begin{bmatrix}
  1\\
  -2\\
  4\\
\end{bmatrix},
v_{2} = \begin{bmatrix}
  2\\
  5\\
  2\\
\end{bmatrix},
v_{3} = \begin{bmatrix}
  1\\
  3\\
  5\\
\end{bmatrix}
$$

```python
import numpy as np

def dotProduct(v1, v2):
  row = np.shape(v1)[0]
  sum = 0
  for i in range(row):
    sum += (v1[i] * v2[i])
  return sum

def checkOrthogonal(*vectors):
  dotProducts = list()
  for i in range(len(vectors)):
    for j in range(i+1, len(vectors)):
      v1 = vectors[i]
      v2 = vectors[j]
      dotProd = dotProduct(v1, v2)
      dotProducts.append(dotProd)
  orthogonal = True
  for i in dotProducts:
    if (i != 0):
      orthogonal = False
      break

  if (orthogonal):
    print("The vectors are orthogonal.")
  else:
    print("The vectors are not orthogonal.")


# declaration and initialization
v1 =  np.array([1, -2, 4])
v2 = np.array([2, 5, 2])
v3 = np.array([1, 3, 5])

# Orthogonal Check
checkOrthogonal(v1, v2, v3)
```

# Task 2
Find an orthonormal basis spanned by a set of vectors:
$$v_{1} = \begin{bmatrix}
  2\\
  -5\\
  1\\
\end{bmatrix},
v_{2} = \begin{bmatrix}
  4\\
  -1\\
  2\\
\end{bmatrix}$$

```python
import math
import numpy as np

v1 = np.array([2, -5, 1], dtype = float)
v2 = np.array([4, -1, 2])

def unitConversion(vec):
  row = np.shape(vec)[0]
  modulus_v = 0
  for i in vec:
    modulus_v += (i**2)
  modulus_v = (modulus_v**0.5)
  for i in range(row):
    vec[i] = vec[i]/float(modulus_v)
  return vec

u1 = unitConversion(v1)

#Run a loop for each item in v1
#Then find square of each item and keep track of the sum
#Find the square root of the sum which will give us the modulus of v1
#Then calculate u1 = v1/mod_v1

#==========================================

w2 = np.zeros(len(v2))
for i in range(len(v2)):
    w2[i] = v2[i] - (np.dot(v2,u1)) * u1[i]

#Run a loop for each item in w2
#Then find square of each item and keep track of the sum
#Find the square root of the sum which will give us the modulus of w2
#Then calculate u2 = w2/mod_w2

u2 = unitConversion(w2)

print("u1 = ", u1)
print("u2 = ", u2)
```
