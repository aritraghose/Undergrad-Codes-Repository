5. $\frac{1}{8}x + \frac{1}{7}y - \frac{1}{6}z  + \frac{1}{5}w= 1$

   $\frac{1}{7}x - \frac{1}{6}y + \frac{1}{5}z  - \frac{1}{4}w= 1$

   $\frac{1}{6}x + \frac{1}{5}y - \frac{1}{4}z  + \frac{1}{3}w= 1$

   $\frac{1}{5}x - \frac{1}{4}y + \frac{1}{3}z  - \frac{1}{2}w= 1$



````python
# General Code for Gauss Elimination and Back Substitution / Solving System of Linear Equations

def solveLinearEquation(mat):
  rows, columns = np.shape(mat)
  ansMat = np.zeros(columns-1)

  for i in range(columns-1):
    for j in range(i+1, rows):
      mat[j, :] = (-(mat[j, i]/mat[i, i])*mat[i, :]) + (mat[j, :])
  for i in range(rows-1, -1, -1):
    ansMat[i] = ((mat[i, -1]) - ((mat[i, 0:columns-1]) @ ansMat)) / mat[i, i]
  return ansMat


````

````python
#Write your code here
# 1

augMat = np.array([[1, 1, 1, 2],
                   [-1, 3, 2, 8],
                   [4, 1, 0, 4]], dtype = float)

print("Augmented Matrix: ")
print(augMat)
print("\n")

ansVec = solveLinearEquation(augMat)
print("Solving the system of linear equations: ")
print(ansVec)

print(f"\nSo, x = {ansVec[0]}, y = {ansVec[1]}, z = {ansVec[2]}")
````
