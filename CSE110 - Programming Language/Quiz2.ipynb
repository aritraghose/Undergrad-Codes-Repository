```python


# without using split()

the_sentence = input("Enter the sentence: ")

the_list = list()
space_string = " "
element = ""
blank_element = ""

for letter in the_sentence:
  if letter != space_string:
    element = element + letter
  elif letter == space_string:
    the_list.append(element)
    element = ""

if element != blank_element:
  the_list.append(element)
print(the_list)

for element in the_list:
  string_to_store = ""
  for i in element:
    string_to_store = i + string_to_store
  print(string_to_store, end=" ")
```
