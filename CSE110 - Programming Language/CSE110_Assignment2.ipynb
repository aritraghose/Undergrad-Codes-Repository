**A binary number can be converted to its corresponding decimal number by multiplying each binary digit with a power of 2 where the power denotes the position of the binary digit**

========================================

$2^{3}$ x 1 + $2^{2}$ x 0 + $2^{1}$ x 1 + $2^{0}$ x 0 = 10


**Sample Input** <br>
1010

10

$2^{3}$ x 1 + $2^{2}$ x 1 + $2^{1}$ x 0 + $2^{0}$ x 1 = 13

1101

13

```python
#Todo

binary_value = int(input("Enter the Binary Value: "))
loop_stopper = 1
decimal_value = 0
power = 0

while (loop_stopper != 0):
  position = binary_value%10
  decimal_value += ((2**power)*position)
  binary_value = binary_value//10
  power += 1
  loop_stopper = binary_value

print(decimal_value)

```

### Task 26
Ask the user for a range (a starting number and an ending number).
**Count** how many numbers are prime numbers and how many numbers are perfect numbers between that range and **print** those numbers.


=========================================================================

**Hint (1):** Declare two strings/lists to store the prime and perfect numbers. Inside the iteration store the values of the prime and perfect numbers in the pre- declared variables.

**Hint (2):**


ans = "Prime numbers: "\
prime_value = 11
ans = ans + str(prime_value) + " "\
print(ans)

*Output:*

=========================================================================

**Example:**between 2 and 6, there are 3 prime numbers (2, 3, 5) and 1 perfect number (6).

**Input:**\
2

6

**Output:**\
Between 2 and 6,\
Found 3 prime numbers \
Found 1 perfect number\
Prime numbers: 2  3  5 \
Perfect numbers: 6



```python
#Todo
starting_number = int(input("Enter the starting number: "))
starting_value = starting_number
ending_number = int(input("Enter the ending number: "))

perfect_numbers = []
prime_numbers = []


while (starting_number <= ending_number):
  the_value = starting_number
  starting_number += 1

  #Checking Pefect number or not

  loop_stopper = 1
  sum_of_divisors = 0

  while (loop_stopper <= the_value):
    the_divisor = loop_stopper
    loop_stopper += 1
    if (the_value%the_divisor == 0) and (loop_stopper == the_value+1):
      sum_of_divisors = sum_of_divisors
    elif (the_value%the_divisor == 0):
      sum_of_divisors += the_divisor

  if sum_of_divisors == the_value:
    perfect_numbers.append(the_value)


  #Checking Prime Number or not

  loop_stopper = 1
  divisor_count = 0

  while (loop_stopper <= the_value):
    the_divisor = loop_stopper
    loop_stopper += 1
    if (the_value%the_divisor == 0):
      divisor_count += 1

  if divisor_count == 2:
    prime_numbers.append(the_value)

# Done checking


print("Between", starting_value, "and", ending_number, ",")

print("Found", len(prime_numbers), "prime numbers")
print("Found", len(perfect_numbers), "perfect numbers")

print("Prime numbers:", end = " ")
for each_value in prime_numbers:
  print(each_value, end=" ")
print("")
print("Perfect numbers:", end = " ")
for each_value in perfect_numbers:
  print(each_value, end=" ")
```
