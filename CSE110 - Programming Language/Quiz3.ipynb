```python
# Answer
the_str = input("Enter the string: ")

new_str = ""
space_char = " "
for i in the_str:
  if i != space_char:
    new_str += i

new_dict = {1: [], 2: []}
for i in new_str:
  i = ord(i)
  if i >= 48 and i <= 57:
    i = chr(i)
    if i not in new_dict[1]:
      new_dict[1].append(i)
  elif i >= 65 and i <= 90:
    i = chr(i)
    i = i.lower()
    if i not in new_dict[2]:
      new_dict[2].append(i)
  elif i >= 97 and i <= 122:
    i = chr(i)
    i = i.lower()
    if i not in new_dict[2]:
      new_dict[2].append(i)


print(new_dict)

```
