```python
# Answer for the Problem 1:

n = int(input("Enter a value: "))

counter = 1
while (counter <= n):
  print("=== Iteration", counter, "===")
  if counter%2 == 1:
    local_counter = 1
    while local_counter <= counter:
      for i in range(1, counter+1):
        print(i, end = "")
      print("")
      local_counter += 1
    print("")
  if counter%2 == 0:
    for i in range(1, counter+1):
      print("X"*i)
    print("")
  counter += 1
```
