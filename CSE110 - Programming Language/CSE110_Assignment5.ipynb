### Task 8
Write a Python program that takes a dictionary as an input from the user and then prints the average of all the values in the dictionary.

 <font color='red'>[You are not allowed to use len() and sum()]</font>

===================================================================

**Hint (1): For taking dictionary input**

Approach(1): For taking dictionary as an input from the user, you may take the whole dictionary as a string using the input() function. Then you can use the split(), strip() functions and conditions to get the keys and values from the string. Finally, you can make the dictionary using the obtained data.

Approach(2): If the first approach seems too difficult you can create an empty dictionary and then just run a simple loop. For each iteration ask the user for a key and a value using the input() function and keep updating the dictionary with the key and value.


**Hint (2):** After you have a dictionary, you can use dictionary functions to get all the values from it, run loop to calculate the sum and the total number of values in the dictionary in order to find out the average.

===================================================================

**Sample Input 1:**

{'Jon': 100, 'Dan':200, 'Rob':300}

**Sample Output 1:**

Average is 200.

===================================================================


**Sample Input 2:**

{'Jon': 100, 'Dan':200, 'Rob':30, 'Ned':110}

**Sample Output 2:**

Average is 110.

===================================================================


```python
#to do

the_string = input("Enter the dictionary: ")

# removing the curly brackets from the string
the_string = the_string.strip("{")
the_string = the_string.strip("}")

# assuming the input string might have spaces after every comma and quotations for each keys
space_char = " "
single_quoation = "'"
double_quoation = '"'

string = ""

for i in the_string:
  if i != space_char and i != single_quoation and i != double_quoation:
    string = string + i

# spliting each pairs
pairs = string.split(",")

#creating the dictionary
the_dict = dict()

for each_pair in pairs:
  new_tup = tuple(each_pair.split(":"))
  key, val = new_tup
  the_dict[key] = val

# dictionary has been created from the string

total_val = 0
key_count = 0

for key,val in the_dict.items():
  value = int(val)
  total_val += value
  key_count += 1

average = int(total_val/key_count)

print("Average is ", average, ".", sep="")
```

&nbsp;&nbsp;&nbsp;&nbsp;

**The formula:**\
**<font color='blue'>*Index value of special character list = (ASCII sum of the word) % (length of special character list)*</font>**


=============================================================================


**Sample Input 1**:\
'I love Programming. Python is love.'\
['@', '$', '&', '#']

**Sample Output 1**:\
Words in the given String: ['I', 'love', 'Programming', 'Python', 'is', 'love']\

From the string 'I love Programming. Python is love.' the {key: value} pair is calculated accordingly:


| Word | ASCII sum|List length|Calculated index|Key|
| --- | --- | --- | --- | --- |

|  |  |  |  |  |
|love| 438|4| 438%4 = 2|'&'|

|Programming| 1155|4| 1155%4 = 3|'#'|
|  |  |  |  |  |

|  |  |  |  |  |
|is| 220|4| 220%4 = 0|'@'|
|  |  |  |  |  |
|love| 438|4| 438%4 = 2|'&'|
|  |  |  |  |  |

=============================================================================

**Explanation 2:**\
From the string 'The secret of getting ahead is getting started.' the {key: value} pair is calculated accordingly:


| Word | ASCII sum|List length|Calculated index|Key|
| --- | --- | --- | --- | --- |
|The| 289|5| 289%5 = 4|'%'|
|  |  |  |  |  |
|secret| 646|5| 646%5 = 1|'='|
|  |  |  |  |  |
|of| 213|5| 213%5 = 3|'*'|
|  |  |  |  |  |
|getting| 754|5| 754%5 = 4|'%'|
|  |  |  |  |  |
|ahead| 499|5| 499%5 = 4|'%'|
|  |  |  |  |  |

|getting| 754|5| 754%5 = 4|'%'|
|  |  |  |  |  |
|started| 759|5| 759%5 = 4|'%'|
|  |  |  |  |  |


The calculation is shown in the given table. Full stop after ‘started.’ is avoided. Also, the word ‘getting’ is a duplicate and therefore it was not inserted again as a value.

=============================================================================


```python
# Todo

string = input("Enter line/lines: ")
the_list = input("Enter the list containing special characters: ")

list_str = ""
space_char = " "
full_stop_char = "."
the_list = the_list.strip("[")
the_list = the_list.strip("]")
single_quo = "'"
double_quo = '"'


for i in the_list:
  if i != space_char and i != single_quo and i != double_quo:
    list_str += str(i)

char_list = list_str.split(",")
# list containing special characters created

new_str = ""
for i in string:
  if i != full_stop_char:
    new_str = new_str + i
  if i == full_stop_char:
    pass

words_list = new_str.split(" ")
#list containing words created

# 1st ouput:
print("Words in the given String:", words_list, sep="")


# creating a dictionary for the second output
final_dict = {}
length = len(char_list)

for char in char_list:
  final_dict[char] = list()

for word in words_list:
  ascii_val = 0
  for i in word:
    ascii_val += ord(i)
  index = ascii_val%length
  char_index = char_list[index]
  if word not in final_dict[char_index]:
    final_dict[char_index].append(word)

# current list may contain empty lists
# but the 2nd example does not contain empty lists

deleting_key_list = []
empty_list = []
for key,val in final_dict.items():
  if val == empty_list:
    deleting_key_list.append(key)

for key in deleting_key_list:
  del final_dict[key]


#2nd output:
print("Answer:", final_dict, sep="")

```
