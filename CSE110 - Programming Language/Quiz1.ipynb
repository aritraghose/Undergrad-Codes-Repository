```python
# Answer

N = int(input("Enter the number of Student ID: "))

counter = 1

while (counter <= N):
  student_ID = int(input("Enter Student ID: "))

  student_ID_str = str(student_ID)

  digit_total = 0



  for each_digit in student_ID_str:
    value = int(each_digit)
    digit_total += value


  if (digit_total >= 5) and (digit_total <= 10):
    print("Ciao")
  elif (digit_total >= 11) and (digit_total <= 15):
    print("Hello")
  elif (digit_total >= 20) and (digit_total <= 25):
    print("Hola")
  elif (digit_total >= 30) and (digit_total <= 35):
    print("Salut")
  else:
    print("ByeBye")

  counter += 1


```
