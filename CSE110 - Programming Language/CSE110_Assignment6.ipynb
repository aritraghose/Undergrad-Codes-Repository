```python

def individul_bonus_calculation(name, earning, goals, bonus):
  total_bonus = 0
  total_bonus = int(goals*(bonus/100)*earning)
  if goals > 30:
    total_bonus += 10000
  elif goals >= 20 and goals <= 30:
    total_bonus += 5000
  print(name, "earned a bonus of", total_bonus, "Taka for", goals, "goals.")


individul_bonus_calculation("Neymar", 1200000, 35, 5)
```
