```python
# Answer for Question 1

the_str = input("Enter the string: ")


list_1 = the_str.split(",")
new_list = []
sum = 0
max = None
min = None

try:
  for i in list_1:
    i = int(i)
    new_list.append(i)
  pre_mod = new_list.copy()
  index = 0
  for i in new_list:
    index += 1
    sum += i
    if max == None and min == None:
      max = i
      min = i
    else:
      if i > max:
        max = i
        max_index = index -1
      elif i < min:
        min = i
        min_index = index - 1

  length = len(new_list)
  average = sum/length
  average = int(average)

  if sum%2 == 0:
    new_list.remove(max)
    new_list.insert(max_index, average)
  if sum%2 == 1:
    new_list.remove(min)
    new_list.insert(min_index, average)

except:
  print("Failed to complete the task!")
  exit()
else:
  print("Before modification:", pre_mod)
  print("After modification:", new_list)

```

```python
# Answer for Question 2

current_sem_data = {"John":[4.0,3.7,4.0], "Tom":[3.3,3.7,4.0], "Harry":[3.3,3.7,4.0]}
previous_sem_data = {"John":[3.88,90], "Tom":[3.52, 70], "Harry":[3.25,85]}


student_name = input("Enter student name: ")
current_total_GPA = 0
for i in current_sem_data[student_name]:
  current_total_GPA += (i*3)

prev_cgpa = previous_sem_data[student_name][0]
completed_credit = previous_sem_data[student_name][1]
total_credit = completed_credit + (3*3)
cgpa = (current_total_GPA + prev_cgpa * completed_credit)/ total_credit

print("Previous semester's completed credits:", completed_credit)
print("Previous semester's CGPA:", prev_cgpa)
print("Total Credits completed including current semester:", total_credit)
print("Current semester's total GPA:", current_total_GPA)
print("Current semester's CGPA:", cgpa)
```

```python
# Answer for Question 3
def det_largest_triangle(N, base_and_height):
  b_and_h_str = base_and_height.split(" ")
  b_and_h = list()
  for i in b_and_h_str:
    b_and_h.append(int(i))
  print("Number of triangles:", N)
  area = 0
  max_area = 0
  total_index = len(b_and_h) - 1
  for i in range(0, total_index, 2):
    area = 0.5 * b_and_h[i] * b_and_h[i+1]
    if area >= max_area:
      base = b_and_h[i]
      height = b_and_h[i+1]
      max_area = area

  print("The area of the largest triangle is", max_area, "with a base of", base, "and a height of", height)




N = int(input("Enter number of triangles: "))
base_and_height = input("Enter base and height")

det_largest_triangle(N, base_and_height)
```
