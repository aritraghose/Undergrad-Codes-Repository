```python
# Problem 1

theString = input("Enter the String: ")

upperCount = 0
lowerCount = 0

for i in theString:
  if "A" <= i <= "Z":
    upperCount += 1
  elif "a" <= i <= "z":
    lowerCount += 1

updatedString = ""

if upperCount > lowerCount:
  for i in theString:
    if "a" <= i <= "z":
      subStr = chr(ord(i)-32)
      updatedString += subStr
    else:
      updatedString += i

else:
  for i in theString:
    if "A" <= i <= "Z":
      subStr = chr(ord(i)+32)
      updatedString += subStr
    else:
      updatedString += i

print(updatedString)
```

```python
# Problem 2

N = int(input("Enter the value of N: "))
counter  = 0
highest = 0
resultantList = list()

while (counter < N):
  givenList = input("Enter List: ")
  givenList = givenList.split(" ")
  total = 0
  for i in givenList:
    i = int(i)
    total += i
  if total > highest:
    highest = total
    resultantList = givenList
  counter += 1

resultList = list()
for i in resultantList:
  i = int(i)
  resultList.append(i)

print(highest)
print(resultList)
```

# **`Dictionary & Tuple`**

```python
# Problem 1

inputDict = input("Enter both the dictionaries separated by a comma: ")

#Since the input contains both the dictionaries seperated by a comma, our input string should look like this:
#inputDict = "a: 100, b: 100, c: 200, d: 300,a: 300, b: 200, d: 400, e: 200"
#if the user seperates the dictionary by a comma and a space, it would have been hard to identify the individual dictioaries.
#for example:
#inputDict = "a: 100, b: 100, c: 200, d: 300, a: 300, b: 200, d: 400, e: 200"
#So, the code is written keeping in mind that the user will enter the the dictionaries the way it is stated in the question and shown in the sample input

sepPoint = None

for i in range(0, len(inputDict)):
  if (inputDict[i] == ",") and ("a" <= inputDict[i+1] <= "z"):
    sepPoint = i

#separating the dictionaries
dict1 = inputDict[0:sepPoint]
dict2 =inputDict[sepPoint+1:]


# writing a function to convert the string to an actual dictionary
def stringToDict(theString):
  listElements = theString.split(", ")
  theDict = {}
  for i in listElements:
    keyValPair = i.split(": ")
    theDict[keyValPair[0]] = int(keyValPair[1])
  return theDict

#passing the values
dict1 = stringToDict(dict1)
dict2 = stringToDict(dict2)
# Conversion done


# code to combine both the dictionaries:

combinedDict = dict()

for key,val in dict1.items():
  combinedDict[key] = val

for key,val in dict2.items():
  if key not in combinedDict:
    combinedDict[key] = val
  else:
    combinedDict[key] += val

# both the dictionaries have been combined
print(combinedDict)

# sorting the unique values and storing in a tuple
uniqueValues = []
for i in combinedDict.values():
  if i not in uniqueValues:
    uniqueValues.append(i)

uniqueValues = tuple(sorted(uniqueValues))

print("Values:", uniqueValues)

```
