```python
# Problem 1

def getFraction(x, y):
  if (x == 0) or (y == 0):
    return 0
  else:
    fraction = (x/y) - (x//y)
    return fraction

fraction = getFraction(5, 2)
print(fraction)
```

```python
# Problem 2

def getBMI(height, weight):
  height = height/100
  BMI = weight/(height**2)
  BMI = round(BMI, 1)
  status = None

  if (BMI < 18.5):
    status = "Underweight"
  elif (18.5 <= BMI <= 24.9):
    status = "Normal"
  elif (25 <= BMI <= 30):
    status = "Overweight"
  else:
    status = "Obese"

  statusDialogue = "Score is " + str(BMI) + ". You are " + status
  return statusDialogue




x = getBMI(175, 96)
y = getBMI(152, 48)

print(x)
print(y)
```

```python
# Problem 3

def divisibleTotal(minimum, maximum, divisor):
  total = 0
  for i in range(minimum, maximum):
    if ((i%divisor) == 0):
      total += i
  return total


x = divisibleTotal(0, 10, 2)
y = divisibleTotal(3, 15, 3)
print(x)
print(y)
```

```python
# Problem 5

def replace_domain(mailAddress, newDomain, oldDomain = "kaaj.com"):
  mailAddress = mailAddress.split("@")
  if (mailAddress[1] != newDomain):
    updatedEmail = mailAddress[0] + "@" + newDomain
    print("Changed:", updatedEmail)
  else:
    updatedEmail = mailAddress[0] + "@" + mailAddress[1]
    print("Unchanged:", updatedEmail)

replace_domain("alice@kaaj.com", "sheba.xyz", "kaaj.com")
replace_domain("bob@sheba.xyz", "sheba.xyz")
```

```python
# Problem 6

def vowelCount(theName):
  vowels = "aeiouAEIOU"
  vowelsFound = list()
  for i in theName:
    if i in vowels:
      vowelsFound.append(i)
  if (len(vowelsFound) == 0):
    print("No vowels in the name")
  else:
    print("Vowels: ", ", ".join(vowelsFound), ". Total number of vowels: ", len(vowelsFound), sep="")


vowelCount("Steve Jobs")
vowelCount("XYZ")
```

```python
# Problem 7

def checkPalindrome(givenSequence):
  theSequence = ""
  for i in givenSequence:
    if (i != " "):
      theSequence += i
  # spaces have been trimmed

  length = int(len(theSequence))
  flag = 1  # we are assuming the given sequence is a palindrome, until it is proven otherwise

  for i in range(0, length//2):
    if theSequence[i] != theSequence[length-i-1]:
      flag = 0

  if (flag == 0):
    print("Not a palindrome")
  else:
    print("Palindrome")



checkPalindrome("madam")
checkPalindrome("hello")
checkPalindrome("nurses run")
```

```python
# Problem 9


def stringCorrection(theString):
  correctedString = list(theString)
  for i in range(0, len(theString)):
    if (i == 0) and ("a" <= theString[i] <= "z"):
      correctedString[i] = chr((ord(theString[i])-32))
      continue
    if (i <= (len(theString)-3)) and theString[i] in "?!.":
      correctedString[i+2] = chr((ord(theString[i+2])-32))
    if (i <= (len(theString)-2)) and (theString[i] == "i"):
      if (theString[i-1] == " ") and (theString[i+1] == " "):
        correctedString[i] = "I"
      elif (theString[i-1] == " ") and (theString[i+1] in "!?."):
        correctedString[i] = "I"

  updatedString = ""
  for i in correctedString:
    updatedString += i
  return updatedString



# test case 1
x = stringCorrection("my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog’s name? i love my pet very much.")
print(x)

# test case 2
y = stringCorrection("How do i know?")
print(y)

# test case 3
z = stringCorrection("i am not getting this, do i?")
print(z)
```
