```python
# Problem 1

class Calculator:
  def __init__(self):
    print("Let’s Calculate!")

  def add(self, value1, value2):
    result = value1+value2
    print("Value 1:", value1)
    print("Operator: +")
    print("Value 2:", value2)
    print("Result:", result)

  def subtract(self, value1, value2):
    result = value1-value2
    print("Value 1:", value1)
    print("Operator: -")
    print("Value 2:", value2)
    print("Result:", result)

  def multiply(self, value1, value2):
    result = value1*value2
    print("Value 1:", value1)
    print("Operator: *")
    print("Value 2:", value2)
    print("Result:", result)

  def divide(self, value1, value2):
    result = value1/value2
    print("Value 1:", value1)
    print("Operator: /")
    print("Value 2:", value2)
    print("Result:", result)


object1 = Calculator()
value1 = int(input("Enter value 1: "))
operator = input("Enter operator: ")
value2 = int(input("Enter value 2: "))

if (operator == "+"):
  object1.add(value1, value2)
elif (operator == "-"):
  object1.subtract(value1, value2)
elif (operator == "*"):
  object1.multiply(value1, value2)
elif (operator == "/"):
  object1.divide(value1, value2)
```

```python
# Problem 2

class Customer():
  def __init__(self, name):
    self.name = name

  def greet(self, name = None):
    if (name == None):
      print("Hello!")
    else:
      print(f"Hello, {self.name}!")

  def purchase(self, *items):
    print(f"{self.name}, you purchased {len(items)} item(s):")
    for item in items:
      print(item)

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
# Problem 3

class Panda:

  def __init__(self, name, gender, age):
    self.name = name
    self.gender = gender
    self.age = age

  def sleep(self, duration = None):
    self.duration = duration
    if (self.duration == None):
      return ("{}'s duration is unknown thus should have only bamboo leaves".format(self.name))
    elif (3 <= self.duration <= 5):
      return ("{} sleeps {} hours daily and should have Mixed Veggies".format(self.name, self.duration))
    elif (6 <= self.duration <= 8):
      return ("{} sleeps {} hours daily and should have Eggplant & Tofu".format(self.name, self.duration))
    elif (9 <= self.duration <= 11):
      return ("{} sleeps {} hours daily and should have Broccoli Chicken".format(self.name, self.duration))



panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
# Problem 5

class Student:
  def __init__(self, name = None):
    if (name == None):
      self.name = "default student"
    else:
      self.name = name

  def quizcalc(self, mark1 = 0, mark2 = 0, mark3 = 0):
    self.average = (mark1+mark2+mark3)/3

  def printdetail(self):
    print(f"Hello {self.name}")
    print(f"Your average quiz score is {self.average}")



s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()
```

```python
# Problem 13

class Batsman:
  def __init__(self, *batsmanStat):
    if (len(batsmanStat) == 3):
      self.name = batsmanStat[0]
      self.runs = batsmanStat[1]
      self.balls = batsmanStat[2]
    elif (len(batsmanStat) == 2):
      self.name = "New Batsman"
      self.runs = batsmanStat[0]
      self.balls = batsmanStat[1]

  def battingStrikeRate(self):
    self.strikeRate = (self.runs/self.balls)*100
    return self.strikeRate

  def setName(self, updatedName):
    self.name = updatedName

  def printCareerStatistics(self):
    print(f"Name: {self.name}")
    print(f"Runs Scored: {self.runs} , Balls Faced: {self.balls}")


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
# Problem 15

class Account:
  def __init__(self, name = "Default Account", balance = 0.0):
    self.name = name
    self.balance = float(balance)

  def details(self):
    return (f"{self.name}\n{self.balance}")

  def withdraw(self, amount):
    if ((self.balance - amount) > 3070):
      print(f"Withdraw successful! New balance is: {self.balance-amount}")
    else:
      print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")


a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)
```

```python
# Problem 16

class Author:
  def __init__(self, *userInput):
    if (len(userInput) == 0):
      self.name = "Default"
      self.bookList = list()
    elif (len(userInput) == 1):
      self.name = userInput[0]
      self.bookList = list()
    elif (len(userInput) > 1):
      self.name = userInput[0]
      self.bookList = list()
      for i in range(1, len(userInput)):
        (self.bookList).append(userInput[i])

  def changeName(self, updatedName):
    self.name = updatedName

  def addBooks(self, *updatedBooks):
    for i in updatedBooks:
      (self.bookList).append(i)


  def printDetails(self):
    print(f"Author Name: {self.name}")
    print(f"--------")
    print(f"List of Books:")
    for i in self.bookList:
      print(i)


auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print("===================")
auth2 = Author()
print(auth2.name)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print("===================")
auth2.printDetails()
print("===================")
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails()
```
