```python
# Task 8
import math

class Cylinder:
  radius = 5
  height = 18
  def __init__(self, radius, height):
    print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
    Cylinder.radius = radius
    Cylinder.height = height
    print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")
  

  @staticmethod
  def area(radius, height):
    area = (2*(math.pi)*(radius**2)) + (2*(math.pi)*radius*height)
    print(f"Area: {area}")
  
  @staticmethod
  def volume(radius, height):
    volume = ((math.pi)*radius*radius*height)
    print(f"Volume: {volume}")

  @classmethod
  def swap(cls, height, radius):
    obj = cls(radius, height)
    return obj
  
  @classmethod
  def changeFormat(cls, inputString):
    radius, height = inputString.split("-")
    obj = cls(float(radius), float(height))
    return obj

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10

class SultansDine:
  totalBranch = 0
  totalSell = 0
  listOfBranches = [] #
  def __init__(self, branch):
    SultansDine.totalBranch += 1
    self.branch = branch
    (SultansDine.listOfBranches).append(self)


  def sellQuantity(self, sell):
    if (sell < 10):
      self.branchSell = sell*300
    elif (sell < 20):
      self.branchSell = sell*350
    else:
      self.branchSell = sell*400

    SultansDine.totalSell += self.branchSell

  def branchInformation(self):
    print(f"Branch Name: {self.branch}")
    print(f"Branch Sell: {self.branchSell} Taka")



  @classmethod
  def details(cls):
    print(f"Total Number of branch(s): {cls.totalBranch}")
    print(f"Total Sell: {cls.totalSell} Taka")
    for i in (cls.listOfBranches):
      print(f"Branch Name: {i.branch}, Branch Sell: {i.branchSell} Taka")
      percentage = ((i.branchSell)/(cls.totalSell))*100
      print(f"Branch consists of total sell's: {round(percentage, 2)}%")



SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```

```python
# Task 5

class Employee:
  def __init__(self, name, workingPeriod):
    self.name = name
    self.workingPeriod = workingPeriod


  @classmethod
  def employeeByJoiningYear(cls, name, year):
    name = name
    workingPeriod = 2022 - int(year)
    obj = cls(name, workingPeriod)
    return obj

  @staticmethod
  def experienceCheck(exp, gender):
    if gender == "male":
      if (exp < 3):
        return ("He is not experienced")
      else:
        return ("He is experienced")
    elif gender == "female":
      if (exp < 3):
        return ("She is not experienced")
      else:
        return ("She is experienced")
    
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
# Task 2

class Assassin:
  totalAssassin = 0
  def __init__(self, name, successRate):
    self.name = name
    self.successRate = successRate
    Assassin.totalAssassin += 1
  
  @classmethod
  def failureRate(cls, name, failureRate):
    successRate = 100-failureRate
    obj = cls(name, successRate)
    return obj
  

  @classmethod
  def failurePercentage(cls, name, failurePercentage):
    name = name
    successRate = 100 - int(failurePercentage[:-1])
    obj = cls(name, successRate)
    return obj


  def printDetails(self):
    print(f"Name: {self.name}")
    print(f"Success rate: {self.successRate}%")
    print(f"Total number of Assassin: {Assassin.totalAssassin}")






john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
# Task 7

class Cat:
  Number_of_cats = 0

  def __init__(self, color, activity):
    self.color = color
    self.activity = activity
    Cat.Number_of_cats += 1
  

  @classmethod
  def no_parameter(cls):
    color = "White"
    activity = "sitting"
    obj = cls(color, activity)
    return obj

  @classmethod
  def first_parameter(cls, color):
    color = color
    activity = "sitting"
    obj = cls(color, activity)
    return obj

  @classmethod
  def second_parameter(cls, activity):
    color = "Grey"
    activity = activity
    obj = cls(color, activity)
    return obj

  def changeColor(self, newColor):
    self.color = newColor
  
  def printCat(self):
    print(f"{self.color} cat is {self.activity}")




print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```
