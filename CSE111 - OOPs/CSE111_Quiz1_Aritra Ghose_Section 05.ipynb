```python
# Answer to Problem number 1

N = int(input("Enter the value of N: "))
orderedDict = {}

for i in range(N):
  playerStats = input("Enter:")
  playerStats = playerStats.split(" ")
  playerStats[2] = int(playerStats[2])
  playerStats[3] = int(playerStats[3])
  playerStats[4] = int(playerStats[4])
  keyTuple = (playerStats[0], playerStats[-1])
  if keyTuple not in orderedDict:
    orderedDict[keyTuple] = {playerStats[1]: [playerStats[2], playerStats[3], playerStats[4]]}
  if keyTuple in orderedDict:
    orderedDict[keyTuple][playerStats[1]] = [playerStats[2], playerStats[3], playerStats[4]]



print(orderedDict)
```
