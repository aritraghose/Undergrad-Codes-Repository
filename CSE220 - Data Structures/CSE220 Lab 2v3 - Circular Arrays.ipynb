```python
from prompt_toolkit.styles.style_transformation import parse_color
from re import T
class CircularArray:
  def __init__(self, lin, st, sz):
    # Initializing Variables
    self.start = st
    self.size = sz
    self.cir = [None]*len(lin)
    
    # if lin = [10, 20, 30, 40, None]
    # then, CircularArray(lin, 2, 4) will generate
    # cir = [40, null, 10, 20, 30]
    
    # To Do. 
    # Hints: set the values for initialized variables
    index = self.start
    for i in range(0, self.size):
      self.cir[index] = lin[i]
      index = (index+1)%len(self.cir)
    


  
  # Print from index 0 to len(cir) - 1
  def printFullLinear(self): #Easy
    for i in range(0, len(self.cir)-1):
      print(self.cir[i], end = ", ")
    print(self.cir[len(self.cir)-1])

  
  # Print from start index and total size elements
  def printForward(self): #Easy
    index = self.start
    for i in range(0, self.size-1):
      print(self.cir[index], end = ", ")
      index = (index+1)%len(self.cir)
    print(self.cir[index])



  
  def printBackward(self): #Easy
    index = ((self.start + self.size)%len(self.cir)) - 1
    for i in range(0, self.size-1):
      print(self.cir[index], end = ", ")
      index = index - 1
      if (index < 0):
        index = len(self.cir) - 1

    print(self.cir[index])
      

  
  # With no null cells
  def linearize(self): #Medium
    newLin = [0]*(self.size)
    index = self.start
    for i in range(0, self.size):
      newLin[i] = self.cir[index]
      index = (index+1)%len(self.cir)
    self.cir = newLin


  
  # Do not change the Start index
  def resizeStartUnchanged(self, newcapacity):
    resizedArray = [None]*newcapacity
    indexOldArray = self.start
    indexNewArray = self.start
    for i in range(0, self.size):
      resizedArray[indexNewArray] = self.cir[indexOldArray]
      indexOldArray = (indexOldArray+1)%len(self.cir)
      indexNewArray = (indexNewArray+1)%newcapacity
    
    self.cir = resizedArray

  
  # This method will check whether the array is palindrome or not
  def palindromeCheck(self): #Hard
    palindrome = True
    i = self.start
    j = (self.start+self.size-1)%len(self.cir)

    for k in range(0, self.size//2):
      if self.cir[i] != self.cir[j]:
        palindrome = False
        break
      i = (i+1)%len(self.cir)
      j = j-1
      if (j == -1):
        j = len(self.cir) - 1
    
    if (palindrome):
      print("This array is a palindrome")
    else:
      print("This array is NOT a palindrome")


    

  # This method will sort the values by keeping the start unchanged
  def sort(self):
    startIndex = self.start
    endIndex = ((self.start + self.size)%len(self.cir))-1 
    while(startIndex != endIndex):
      start = startIndex
      end = endIndex
      while (start != end):
        next = (start+1)%len(self.cir)
        if (self.cir[start] > self.cir[next]):
          temp = self.cir[next]
          self.cir[next] = self.cir[start]
          self.cir[start] = temp
        start = (start+1)%len(self.cir)


      endIndex = endIndex - 1
      if (endIndex < 0):
        endIndex = len(self.cir) - 1
      



    
  
  # This method will check the given array across the base array and if they are equivalent interms of values return true, or else return false
  def equivalent(self, cir_arr):
    equivalentArray = True
    i = self.start
    j = cir_arr.start
    if (self.size != cir_arr.size):
      return False
    else:
      for k in range(self.size):
        if (self.cir[i] != cir_arr.cir[j]):
          equivalentArray = False
          break
        i = (i+1)%len(self.cir)
        j = (j+1)%len(cir_arr.cir)
      
    return equivalentArray
      

  # the method take another circular array and returns a linear array containing the common elements between the two circular arrays.
  def intersection(self, c2):
    newArr = [None]*self.size
    p = 0 # iterator of the new array with common elements

    m = self.start  # iterator for array 1
    for i in range(self.size):
      n = c2.start # iterator for array 2
      for j in range(c2.size):
        if (self.cir[m] == c2.cir[n]):
          newArr[p] = self.cir[m]
          c2.cir[n] = None
          p = p+1
          break
        n = (n+1)%len(c2.cir)
      m = (m+1)%len(self.cir)

    return newArr[0:p]



```
