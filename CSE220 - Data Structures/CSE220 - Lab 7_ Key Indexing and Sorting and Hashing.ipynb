```python
# Task 1

class KeyIndex:
  def __init__ (self, a):
    min = a[0]
    max = a[0]
    for i in range(1, len(a)):
      if (a[i] < min):
        min = a[i]
      if (a[i] > max):
        max = a[i]
    
    size = (max + 1)
    self.shift = 0
    if (min < 0):
      self.shift = -1*(min)
      size = size - min
    self.k = [0]*size
    
    for i in range(len(a)):
      self.k[a[i] + self.shift] += 1


  def search(self, val):
    if (self.k[val+(self.shift)] == 0):
      return False
    return True
  
  def sort(self):
    newArr = [0]*len(self.k)
    i = 0
    for j in range(len(self.k)):
      for k in range(self.k[j]):
        newArr[i] = j - self.shift
        i += 1
    return newArr[0:i]



# Tester Class
class TesterClass:
  def __init__(self, arr):
    self.arr = arr
    self.keyList = KeyIndex(arr)

  def checkSearchFunction(self):
    print("Search Function: ")
    print("Searching -5: ", self.keyList.search(-5))
    print("Searching -9: ", self.keyList.search(-9))

  def checkSortingFunction(self):
    print("\nSorting Function: ")
    print("Before Sorting: ", self.arr)
    print("After Sorting: ", self.keyList.sort())


# Driver Code
print("Task 1: ")

arr = [5, 9, -5, -6, 15]
test1 = TesterClass(arr)
test1.checkSearchFunction()
test1.checkSortingFunction()
```

```python
# Task 2

class Hashtable():
  def __init__(self, a):
    self.hashtable = [None]*len(a)
    for i in range(len(a)):
      the_str = a[i]
      # Hash function calculation
      alphaCal = 0
      numCal = 0
      for j in range(len(the_str)):
        if (48 <= ord(the_str[j]) <= 57):
          numCal += int(the_str[j])
          continue
        elif (65 <= ord(the_str[j]) <= 90) and the_str[j] != 'A' and the_str[j] != 'E' and the_str[j] != 'I' and the_str[j] != 'O' and the_str[j] != 'U':
          alphaCal += 1
          continue
      hashCalc = ((alphaCal*24) + numCal) % len(a)
      #print(alphaCal, numCal, hashCalc)


      if (self.hashtable[hashCalc] == None):
        self.hashtable[hashCalc] = the_str
      else:
        # Linear Probing
        idx = (hashCalc+1)%len(self.hashtable)
        for i in range(len(self.hashtable)):
          if (self.hashtable[idx] == None):
            self.hashtable[idx] = the_str
            break
          else:
            idx = (idx+1)%len(self.hashtable)



# Driver Code
test_str = ["ST1E89B8A32", "DF465FDH45", "EW8974HNRTGH", "SJ465RH78", "FS45DOG6G", "SG46BHET87", "78121345", "SFGERHTEHT", "AAA111BB22"]
t1 = Hashtable(test_str)
print(t1.hashtable)
```
