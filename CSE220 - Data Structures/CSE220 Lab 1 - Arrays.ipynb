```python
# task 1

def shiftLeft(source, k):
  for i in range(k, len(source)):
    source[i-k] = source[i]
  for i in range(len(source)-k, len(source)):
    source[i] = 0


source = [10,20,30,40,50,60]
shiftLeft(source,2)
print(source)
```

```python
# task 2
def rotateLeft(source, k):
  l = 0
  for i in range(k, len(source)):
    for j in range(i, l, -1):
      temp = source[j-1]
      source[j-1] = source[j]
      source[j] = temp
    l = l+1


source = [10,20,30,40,50,60]
rotateLeft(source,5)
print(source)
```

```python
# task 5

def remove(source, size, idx):
  for i in range(idx+1, size):
    source[i-1] = source[i]
  source[size-1] = 0
  size = size-1


source=[10,20,30,40,50,0,0]
remove(source,5,2)
print(source)
```

```python
# Task 7

def checkWeightBalance3(source):
  balance = False
  for mid in range (0, len(source)-1):
    leftTotal = 0
    for i in range(0, mid+1):
      leftTotal += source[i]
    rightTotal = 0
    for j in range(mid+1, len(source)):
      rightTotal += source[j]
    if (leftTotal == rightTotal):
      balance = True
      return True
  return balance


source =  [8, 7, 1, 0, 4, 10]
print(checkWeightBalance3(source))
```

```python
# Task 7 v2

def checkWeightBalance(source):
  balance = False
  for mid in range (0, len(source)+1):
    leftTotal = 0
    for i in range(0, mid):
      leftTotal += source[i]
    rightTotal = 0
    for j in range(mid, len(source)):
      rightTotal += source[j]
    if (leftTotal == rightTotal):
      balance = True
      return True
  return balance


source =  [10, 3, 1, 2, 10]
print(checkWeightBalance(source))
```

```python
# Task 8

def createSeries(n):
  arr = [0]*(n*n)
  m = 1
  for i in range(n-1, (n*n)+1, n):
    element = 1
    for j in range(0, m):
      arr[i-j] = element
      element += 1
    m += 1
  return arr

arr = createSeries(5)
print(arr)
```

```python
# Task 9

def maxBunchCount(source):
  maxCount = 1
  i = 1
  while (i < len(source)):
    count = 1
    while (source[i] == source[i-1]):
      count += 1
      i += 1
      if (i == len(source)):
        break
    if (count > maxCount):
      maxCount = count
    i += 1
  return maxCount

arr = [1,1,2, 2, 2, 2, 2, 1, 1,1,1]
print(maxBunchCount(arr))
```

```python
# Task 10

def sameRepetitionCheck(source):
  repetitionCounter = [0]*len(source)
  m = 0
  for i in range(0, len(source)):
    count = 1
    for j in range(i+1, len(source)):
      if ((source[i] != 0) and (source[i] == source[j])):
        count += 1
        source[j] = 0
    if (count > 1):
      repetitionCounter[m] = count
      m += 1


  for i in range(0, m):
    for j in range(i+1, m):
      if (repetitionCounter[i] == repetitionCounter[j]):
        return True
  return False



arr = [3,5,6,6, 1, 1, 5, 5, 3]
print(sameRepetitionCheck(arr))
```
