```python
# Task 1(a)
def factorial(n):
  if (n == 0):
    return 1;
  return (n * factorial(n-1))


# Task 1(b)
def fib(n):
  if (n == 0) or (n == 1):
    return n 
  return fib(n-1) + fib(n-2)


# Task 1(c)
def printElemenets(arr, i=0):
  if (i == len(arr)):
    return
  print(arr[i], end = " ")
  printElemenets(arr, i+1)


# Task 1(d)
def baseToThePower(base, n):
  if (n == 0):
    return 1
  return (base * baseToThePower(base, n-1))


# Driver Code
# Task 1(a)
print("Task 1(a): ")
print("0! = ", factorial(0))
print("6! = ", factorial(6))

# Task 1(b)
print("\nTask 1(b): ")
print("First 10 fibonacci numbers: ")
for i in range(10):
  print(fib(i))


# Task 1(c)
print("\nTask 1(c): ")
arr = [1, 2, 3, 4, 5]
printElemenets(arr)
print()

# Task 1(d)
print("\nTask 1(d): ")
print("2^4 = ", baseToThePower(2, 4))
print("5^0 = ", baseToThePower(5, 0))
print("6^6 = ", baseToThePower(6, 6))
print("0^1 = ", baseToThePower(0, 1))
```

```python
# Task 2(a)

def decToBin(n):
  if (n == 0):
    return "0"
  smallPart = n%2
  return decToBin(n//2) + str(smallPart)


print(decToBin(1))
print(decToBin(32))
print(decToBin(35))
```

```python
# Task 2(b)

def sum(current):
  if (current == None):
    return 0
  return current.element + sum(current.next)


# Task 2(c)
def printBackward(current):
  if current == None:
    return
  printBackward(current.next)
  print(current.element)


# Driver Code

class Node():
  def __init__(self, elem, next):
    self.element = elem
    self.next = next

class CreateLinkedList:
  def __init__(self, arr):
    self.head = None
    tail = None
    for i in range(len(arr)):
      if (self.head == None):
        newNode = Node(arr[i], None)
        self.head = newNode
        tail = newNode
      else:
        newNode = Node(arr[i], None)
        tail.next = newNode
        tail = tail.next
  
  def returnHead(self):
    return self.head


arr = [10, 20, 30, 40, 50, 60]
l1 = CreateLinkedList(arr)
head = l1.returnHead()

# Task 2(b)
print("Task 2(b): ")
sum = sum(head)
print(sum)

# Task 2(c)
print("\nTask 2(c)")
printBackward(head)
```

```python
# Task 3

def hocBuilder(height):
  if (height == 0):
    return 0
  if (height == 1):
    return 8
  return 5 + hocBuilder(height-1)


# Driver Code
print(hocBuilder(5))
print(hocBuilder(1))
print(hocBuilder(0))
```

```python
# Task 4(a)

def printer(n):
  if (n == 1):
    print(n, end= " ")
    return
  printer(n-1)
  print(n, end = " ")


def printPattern(n):
  if (n == 1):
    printer(n)
    return
  printPattern(n-1)
  print()
  printer(n)



# Driver Code
printPattern(5)

print("\n")
printPattern(6)
```
