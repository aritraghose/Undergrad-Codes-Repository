```python
# Task 1(a)
def factorial(n):
  if (n == 0):
    return 1;
  return (n * factorial(n-1))


# Task 1(b)
def fib(n):
  if (n == 0) or (n == 1):
    return n 
  return fib(n-1) + fib(n-2)


# Task 1(c)
def printElemenets(arr, i=0):
  if (i == len(arr)):
    return
  print(arr[i], end = " ")
  printElemenets(arr, i+1)


# Task 1(d)
def baseToThePower(base, n):
  if (n == 0):
    return 1
  return (base * baseToThePower(base, n-1))


```

```python
# Task 2(b)

def sum(current):
  if (current == None):
    return 0
  return current.element + sum(current.next)


# Task 2(c)
def printBackward(current):
  if current == None:
    return
  printBackward(current.next)
  print(current.element)




class Node():
  def __init__(self, elem, next):
    self.element = elem
    self.next = next

class CreateLinkedList:
  def __init__(self, arr):
    self.head = None
    tail = None
    for i in range(len(arr)):
      if (self.head == None):
        newNode = Node(arr[i], None)
        self.head = newNode
        tail = newNode
      else:
        newNode = Node(arr[i], None)
        tail.next = newNode
        tail = tail.next
  
  def returnHead(self):
    return self.head


arr = [10, 20, 30, 40, 50, 60]
l1 = CreateLinkedList(arr)
head = l1.returnHead()


sum = sum(head)
print(sum)

printBackward(head)
```

```python
# Task 3

def hocBuilder(height):
  if (height == 0):
    return 0
  if (height == 1):
    return 8
  return 5 + hocBuilder(height-1)



print(hocBuilder(5))
print(hocBuilder(1))
print(hocBuilder(0))
```

```python
# Task 4(a)

def printer(n):
  if (n == 1):
    print(n, end= " ")
    return
  printer(n-1)
  print(n, end = " ")


def printPattern(n):
  if (n == 1):
    printer(n)
    return
  printPattern(n-1)
  print()
  printer(n)




printPattern(5)

print("\n")
printPattern(6)
```
