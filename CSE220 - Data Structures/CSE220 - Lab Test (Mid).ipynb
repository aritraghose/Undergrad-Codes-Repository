```python
# Answer


# Task 1
class Node():
  def __init__(self, element, next):
    self.element = element
    self.next = next



class LinkedList():
  def __init__(self, arr):
    self.head = None
    tail = None
    for i in range(len(arr)):
      newNode = Node(arr[i], None)
      if (self.head == None):
        self.head = newNode
        tail = newNode
      else:
        tail.next = newNode
        tail = tail.next

  def printForward(self):
    current = self.head
    while (current.next != None):
      print(current.element, end = ",")
      current = current.next
    print(current.element)

  def reverse(self):
    current = self.head
    oldHead = None
    while (current != None):
      next = current.next
      self.head = current
      self.head.next = oldHead
      oldHead = self.head
      current = next




  def merge(self, secondHead):
    current = secondHead
    second = self.head
    self.head = secondHead
    i = 0
    while (current.next != None):
      next = current.next
      current.next = second
      second = second.next
      current.next.next = next
      current = next
      i += 1
    while (second != None):
      current.next = second
      second = second.next






```

```python
list1=[40,30,20,10]
l1=LinkedList(list1)
print("-----------------------------")
l1.printForward()
print("-----------------------------")
l1.reverse()
l1.printForward()
print("-----------------------------")
list2=[100,110,120,130]
l2=LinkedList(list2)
l2.merge(l1.head)
l2.printForward()
print("-----------------------------")
```
