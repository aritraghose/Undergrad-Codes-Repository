```python
def trianglePrinter(the_str, op =""):
  if (len(the_str) == 0):
    return (op + the_str)
  op += the_str
  if (len(the_str) > 1):
    op += "\n"
  the_str = the_str[0:len(the_str)-1]
  return trianglePrinter(the_str, op)




def alphaCheck(the_str, i = 0, opStr = ""):
  # base case
  if (len(the_str) == i):
    if (opStr == ""):
      return "Alpha"
    else:
      return trianglePrinter(opStr)
      
  
  if (i == 0):
    return alphaCheck(the_str, i+1, opStr)
  else:
    if (the_str[i] > the_str[i-1]):
      return alphaCheck(the_str, i+1, opStr)
    else:
      opStr += the_str[i]
      return alphaCheck(the_str, i+1, opStr)

print("Test Case 1:")
print(alphaCheck("mnop"))
print("Test Case 2:")
print(alphaCheck("mnopbba"))
print("Test Case 3:")
print(alphaCheck("abcde"))
print("Test Case 4:")
print(alphaCheck("bba"))
```
