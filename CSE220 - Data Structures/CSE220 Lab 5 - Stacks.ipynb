```python
# Task 1

class ArrayStack:
  def __init__(self, size):
    self.stack = [None]*size
    self.size = 0
  def push(self, element):
    self.stack[self.size] = element
    self.size += 1
  def peek(self):
    if (self.size == 0):
      return "Empty Stack"
    return self.stack[self.size-1]
  def pop(self):
    poppedElement = self.stack[self.size-1]
    self.stack[self.size-1] = None
    self.size -= 1
    return poppedElement

  def isEmpty(self):
    if self.size == 0:
      return True
    else:
      return False



def parenthesisBalanceCheck(eqn):
  valid = True
  announced = False
  stack = ArrayStack(len(eqn))
  pos_stack = ArrayStack(len(eqn))
  for i in range(len(eqn)):
    if (not valid):
      break
    if (eqn[i] == '(') or (eqn[i] == '{') or (eqn[i] == '['):
      stack.push(eqn[i])
      pos_stack.push(i+1)
    elif (eqn[i] == ')') or (eqn[i] == '}') or (eqn[i] == ']'):
      if stack.isEmpty():
        pos_stack.push(i+1)
        valid = False
        print("This expression is NOT correct.")
        print(f"Error at character # {pos_stack.peek()}. '{eqn[i]}'- not opened.")
        announced = True
        break
      else:
        popped_parenthesis = stack.pop()
        if ((eqn[i] ==  ')') and (popped_parenthesis != '(')) or ((eqn[i] ==  '}') and (popped_parenthesis != '{')) or ((eqn[i] ==  ']') and (popped_parenthesis != '[')):
          valid = False
          print("This expression is NOT correct.")
          print(f"Error at character # {pos_stack.peek()}. '{popped_parenthesis}'- not closed.")
          announced = True
          break
        else:
          pos_stack.pop()



  if (not announced):
    if stack.isEmpty():
      print("This expression is correct.")
    else:
      print("This expression is NOT correct.")
      print(f"Error at character # {pos_stack.peek()}. '{stack.peek()}'- not closed.")

eqn = input("Enter expression: ")
parenthesisBalanceCheck(eqn)
```

```python
# Task 2

class Node:
  def __init__(self, element, next):
    self.element = element
    self.next = next

class LinkedListStack:
  def __init__(self):
    self.head = None
  def push(self, element):
    newNode = Node(element, None)
    if (self.head == None):
      self.head = newNode
    else:
      newNode.next = self.head
      self.head = newNode
  def peek(self):
    if (self.head == None):
      return "Empty Stack"
    return self.head.element
  def pop(self):
    current = self.head
    self.head = self.head.next
    element = current.element
    current.element = None
    current.next = None
    return element
  def isEmpty(self):
    if (self.head == None):
      return True
    else:
      return False


def parenthesisBalanceCheck(eqn):
  valid = True
  announced = False
  stack = LinkedListStack()
  pos_stack = LinkedListStack()
  for i in range(len(eqn)):
    if (not valid):
      break
    if (eqn[i] == '(') or (eqn[i] == '{') or (eqn[i] == '['):
      stack.push(eqn[i])
      pos_stack.push(i+1)
    elif (eqn[i] == ')') or (eqn[i] == '}') or (eqn[i] == ']'):
      if stack.isEmpty():
        pos_stack.push(i+1)
        valid = False
        print("This expression is NOT correct.")
        print(f"Error at character # {pos_stack.peek()}. '{eqn[i]}'- not opened.")
        announced = True
        break
      else:
        popped_parenthesis = stack.pop()
        if ((eqn[i] ==  ')') and (popped_parenthesis != '(')) or ((eqn[i] ==  '}') and (popped_parenthesis != '{')) or ((eqn[i] ==  ']') and (popped_parenthesis != '[')):
          valid = False
          print("This expression is NOT correct.")
          print(f"Error at character # {pos_stack.peek()}. '{popped_parenthesis}'- not closed.")
          announced = True
          break
        else:
          pos_stack.pop()



  if (not announced):
    if stack.isEmpty():
      print("This expression is correct.")
    else:
      print("This expression is NOT correct.")
      print(f"Error at character # {pos_stack.peek()}. '{stack.peek()}'- not closed.")

eqn = input("Enter expression: ")
parenthesisBalanceCheck(eqn)

```
