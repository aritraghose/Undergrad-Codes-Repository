```python
input_file = open("/content/sample_data/task(B).txt", "r")
output_file = open("/content/sample_data/outputB.txt", "w")

networks = input_file.readline().strip().split(",")

connections_main = [i.strip() for i in input_file.readlines()]
helper = []
network_proxy = []
for i in range(len(networks)):
  network_proxy.append(i)
  helper.append(i)
  helper[i] = False

parent = [i for i in range(len(networks))]
rank = [1 for i in range(len(networks))]

def root(i):
  if (parent[i] != i):
    parent[i] = root(parent[i])
  return parent[i]


def join(u, v):
  parent_u = root(u)
  parent_v = root(v)

  if (parent_u == parent_v):
    return
  if (rank[parent_u] > rank[parent_v]):
    parent[parent_v] = parent_u
  elif (rank[parent_u] < rank[parent_v]):
    parent[parent_u] = parent_v
  else:
    parent[parent_v] = parent_u
    rank[parent_u] += 1


for i in range(len(connections_main)):
  u, v = connections_main[i].strip().split(" ")
  x = ord(u[0]) - 96
  y = ord(v[0]) - 96
  helper[x] = True
  helper[y] = True
  join(x, y)


count = 0

for i in range(1, len(parent)):
  x = root(i)
  if i == x and helper[i] == True:
    count += 1


output_file.write(f"Total networks: {count}\nNetwork connections: [")
str = "'',"
for i in range(len(parent)):
  if i == 0:
    continue
  if helper[i] == False:
    str += f"'Disconnected',"
  else:
    if i == root(i):
      str += f"'central Router({networks[i]})',"
    else:
      str += f"'{networks[i]}',"
str += "]"
output_file.write(str)
output_file.close()
```
