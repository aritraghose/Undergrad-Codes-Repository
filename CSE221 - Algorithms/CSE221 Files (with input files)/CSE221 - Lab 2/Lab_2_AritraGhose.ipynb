```python
# Task 1(b): O(nlogn)

# using two pointer

input_file = open("/content/sample_data/input1_4.txt", mode = "r")
input_data = input_file.readlines()



n_and_s = input_data[0].strip().split(" ")
n = int(n_and_s[0])
s = int(n_and_s[1])
arr = [int(i) for i in (input_data[1]).split(" ")]


output_file = open('/content/sample_data/output1_5_alt.txt', mode = 'w')


# answer part begins

# merge sort with index tracking enabled

def merge(a, b, a_i, b_i):
  i = 0
  j = 0
  k = 0
  newArr = [0]*(len(a) + len(b))
  new_I = [0]*(len(newArr))
  m = 0
  while (i != len(a) and j != len(b)):
    if (a[i] < b[j]):
      newArr[k] = a[i]
      new_I[m] = a_i[i]
      m += 1
      i += 1
      k += 1
    else:
      newArr[k] = b[j]
      new_I[m] = b_i[j]
      m += 1
      j += 1
      k += 1

  while (i != len(a)):
    newArr[k] = a[i]
    new_I[m] = a_i[i]
    m += 1
    i += 1
    k += 1
  while (j != len(b)):
    newArr[k] = b[j]
    new_I[m] = b_i[j]
    j += 1
    k += 1
    m += 1

  return [newArr, new_I]

def mergeSort(arr, index_tracker):
  if (len(arr) <= 1):
    return [arr, index_tracker]
  else:
    mid = len(arr)//2
    a1 = mergeSort(arr[0:mid], index_tracker[0:mid])
    a2 = mergeSort(arr[mid:], index_tracker[mid:])
    return merge(a1[0], a2[0], a1[1], a2[1])



index_tracker = [i for i in range(1, n+1)]
sortedArr = mergeSort(arr, index_tracker)
arr = sortedArr[0]
indices = sortedArr[1]


l = 0
r = len(arr)-1

flag = False
while (l < r):
  if (arr[l] + arr[r] == s):
    # maintaining output convention
    print(arr[l]+arr[r])
    if (indices[l] < indices[r]):
      output_file.write(f"{indices[l]} {indices[r]}")
    else:
      output_file.write(f"{indices[r]} {indices[l]}")
    flag = True
    break
  elif (arr[l]+arr[r] < s):
    l += 1
  else:
    r -= 1

if (flag == False):
  output_file.write(f"IMPOSSIBLE")


output_file.close()
```
