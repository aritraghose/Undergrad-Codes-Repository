```python
# Task 2

# Implementation 1
def fibonacci_1(n):
  if n < 0:
    print("Invalid input!")
  elif n <= 1:
    return n
  else:
    return fibonacci_1(n-1)+fibonacci_1(n-2)

# Implementation 2
def fibonacci_2(n):
  if n<0:
    return "Invalid Input"
  if n<=1:
    return n
  fib = [0] * (n+1)
  fib[0] = 0
  fib[1] = 1
  for i in range(2,n+1):
    fib[i] = fib[i-1] + fib[i-2]
  return fib[n]



import time
import math
import matplotlib.pyplot as plt
import numpy as np

#change the value of n for your own experimentation
n = 30
x = [i for i in range(n)]
y = [0 for i in range(n)]
z = [0 for i in range(n)]


for i in range(n-1):
  start = time.time()
  fibonacci_1(x[i+1])
  y[i+1]= time.time()-start
  start = time.time()
  fibonacci_2(x[i+1])
  z[i+1]= time.time()-start


x_interval = math.ceil(n/10)
plt.plot(x, y, 'r')
plt.plot(x, z, 'b')
plt.xticks(np.arange(min(x), max(x)+1, x_interval))
plt.xlabel('n-th position')
plt.ylabel('time')
plt.title('Comparing Time Complexity!')
plt.show()


# We can see here, for implementation 1, the time taken by the program to generate n_th fibonacci number is getting higher exponentially as n increases. The time complexity for the program is O(2^n)
# But, for implementation 2, with memoization, the time complexity for the program is O(n) - which is more efficient.
```

```python
# Task 3

input_file = open('/content/sample_data/input3.txt', mode = 'r')
input_data = input_file.readlines()
size = int(input_data[0])

arr = input_data[1].split(" ")
for i in range(size):
  arr[i] = int(arr[i])


def bubbleSort(arr):
  for i in range(len(arr)-1):
    flag = False
    for j in range(len(arr)-i-1):
      if arr[j] > arr[j+1]:
        flag = True
        arr[j], arr[j+1] = arr[j+1], arr[j]
    if flag == False:
      break

bubbleSort(arr)

# here, in this optimized version of the Bubble Sort Algorithm, we use a flag to check if there are any further requirement of iterations or not
# for example, lets assume an array: 2, 3, 1, 5, 6
# after the first two outer loop iteration, the array becomes this: 1, 2, 3, 5, 6
# after this point, our only concern is this portion of the array: 1, 2, 3        [since, after two iterations two biggest elements are already placed at the right positions]
# now during this, third iteration, our flag is initialized as False. Then the inner loop begins..
# during the iterations of the inner loop, no elements are swapped. So, the flag remains False as initialized
# Now, since at the end of the inner loop, we see there were no swappings required, meaning all are rightly placed, we break out of the loop.
# We can say that, since 1, 2, 3 are rightly placed, during the next iteration our concerning part of the array: 1, 2 will also be rightly placed
# This is how we use a flag to optimize the Bubble Sort Algorithm
# In short, we check if there were any requirement of swappings in the inner loop or not.
# If, there were no swappings required, that means, there will be no requirement of any swapping in the further iterations.


output_file = open('/content/sample_data/output3.txt', mode = 'w')
for i in range(size):
  output_file.write(f"{arr[i]} ")

output_file.close()
```

```python
# Task 4

# Selection Sort requires minimum number of swappings in order to sort an array

input_file = open("/content/sample_data/input4.txt", mode = "r")
input_data = input_file.readlines()
size = int(input_data[0])
id_list = [int(i) for i in (input_data[1].split(" "))]
marks = [int(i) for i in (input_data[2].split(" "))]

# sorting is done in descending order
for i in range(size):
  max = marks[i]
  max_index = i
  for j in range(i, size):
    if (marks[j] > max):
      max = marks[j]
      max_index = j
    elif (marks[j] == max):
      if id_list[j] <= id_list[max_index]:
        max_index = j
  if (max_index != i):
    marks[i], marks[max_index] = marks[max_index], marks[i]
    id_list[i], id_list[max_index] = id_list[max_index], id_list[i]



output_file = open('/content/sample_data/output4.txt', mode = 'w')

for i in range(size):
  output_file.write(f"ID: {id_list[i]} Mark: {marks[i]}\n")

output_file.close()
```

```python
# Task 5

input_file = open("/content/sample_data/input5.txt", mode = "r")
input_data = input_file.readlines()

n = int(input_data[0])
for i in range(1, n+1):
  input_data[i] = (input_data[i].strip()).split(" ")
trains = [input_data[i][0] for i in range(1, n+1)]
dest = [input_data[i][4] for i in range(1, n+1)]
time = [input_data[i][6] for i in range(1, n+1)]
serial = [i for i in range(1, n+1)]


for i in range(n):
  min = trains[i]
  min_index = i
  for j in range(i, n):
    if (trains[j] < min):
      min = trains[j]
      min_index = j
    elif (trains[j] == min):
      temp = time[j].split(":")
      j_time = ((int(temp[0]))*60) + int(temp[1])
      temp = time[min_index].split(":")
      min_time = ((int(temp[0]))*60) + int(temp[1])
      if (j_time > min_time):
        min_index = j
      elif (j_time == min_time):
        if (serial[j] <= serial[min_index]):
          min_index = j
  if (min_index != i):
    trains[i], trains[min_index] = trains[min_index], trains[i]
    dest[i], dest[min_index] = dest[min_index], dest[i]
    time[i], time[min_index] = time[min_index], time[i]
    serial[i], serial[min_index] = serial[min_index], serial[i]



output_file = open('/content/sample_data/output5.txt', mode = 'w')

for i in range(n):
  output_file.write(f"{trains[i]} will departure for {dest[i]} at {time[i]}\n")

output_file.close()
```
