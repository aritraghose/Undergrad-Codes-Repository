```python
import heapq

h = dict()
g = dict()
adjList = dict()

def buildGraph(textFile):
  for line in textFile:
    line = line.strip()
    chunks = line.split(" ")
    h[chunks[0]] = int(chunks[1])
    adjList[chunks[0]] = []
    for i in range(2, len(chunks), 2):
      adjList[chunks[0]].append([chunks[i], int(chunks[i+1])])



def aStarSearch(start, goal):
  if (start not in adjList) or (goal not in adjList):
    return ({}, -1)
  pq = []
  visited = []
  parent = {}
  heapq.heapify(pq)
  heapq.heappush(pq, (h[start], start, None, 0))

  while(len(pq) > 0):
    weight, node, p, cost = heapq.heappop(pq)
    visited.append(node)
    parent[node] = p
    if node == goal:
      return (parent, cost)

    for neighbour in adjList[node]:
      if neighbour[0] in visited:
        continue
      calc = cost + neighbour[1] + h[neighbour[0]]
      heapq.heappush(pq, (calc, neighbour[0], node, cost+neighbour[1]))
      parent[neighbour[0]] = node

  return ({}, -1)







# Driver Code

input_file = open("/content/sample_data/Input file.txt", "r")
input_data = input_file.readlines()
buildGraph(input_data)

start = input("Start Node: ")
goal = input("Destination: ")
parent, cost = aStarSearch(start, goal)

if cost == -1:
  print("NO PATH FOUND")
else:
  path = ""
  current = goal
  while(current != None):
    path = " -> " + current + path
    current = parent[current]
  path = "Path: " + path[4:]

  print(path)
  print(f"Total Distance: {cost} km")
```
