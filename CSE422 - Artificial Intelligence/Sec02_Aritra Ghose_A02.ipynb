```python
import random
input_file = open("/content/sample_data/Input3.txt", "r")
input_data = input_file.readlines()
player_data = list()
batsmenCount = None
target = None

for i in range(len(input_data)):
  if (i == 0):
    batsmenCount, target = input_data[i].strip().split(" ")
    batsmenCount = int(batsmenCount)
    target = int(target)
    continue
  name, runs = input_data[i].strip().split(" ")
  player_data.append([name, int(runs)])


def fitnessFunction(chromosome):
  runs = 0
  for i in range(len(chromosome)):
    if chromosome[i]:
      runs += player_data[i][1]
  return runs

def mutationFunction(child):
  randomGene = random.randint(0, len(child)-1)
  if child[randomGene]:
    child[randomGene] = 0
  else:
    child[randomGene] = 1
  return child

def crossoverFunction(chromosome1, chromosome2):
  intersectionPoint = random.randint(2, len(chromosome1)-2)
  newChild1 = []
  newChild2 = []
  for i in range(0, len(chromosome1)):
    if i < intersectionPoint:
      newChild1.append(chromosome1[i])
      newChild2.append(chromosome2[i])
    else:
      newChild1.append(chromosome2[i])
      newChild2.append(chromosome1[i])
  return (newChild1, newChild2)


def GeneticAlgorithm(data, size, target):
  population = []
  for i in range(size):
    temp = []
    for j in range(size):
      x = random.randint(0, 1)
      temp.append(x)
    population.append(temp)

  iterations = 1e5
  while (iterations):
    iterations -= 1
    for chromosome in population:
      fitnessVal = fitnessFunction(chromosome)
      if fitnessVal == target:
        return chromosome

    nextGenPopulation = []
    for i in range(0, len(population), 2):
      if i+1 >= len(population):
        continue
      child1, child2 = crossoverFunction(population[i], population[i+1])
      mutatedChild1 = mutationFunction(child1)
      mutatedChild2 = mutationFunction(child2)
      nextGenPopulation.append(mutatedChild1)
      nextGenPopulation.append(mutatedChild2)
    population = nextGenPopulation
  return [-1]







op = GeneticAlgorithm(player_data, batsmenCount, target)
ans = "["
for i in range(len(player_data)):
  if (i == 0):
    ans += f"'{player_data[i][0]}'"
  else:
    ans += f", '{player_data[i][0]}'"
ans += "]\n"


for i in op:
  ans += str(i)

print(ans)
```
