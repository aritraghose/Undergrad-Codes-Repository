```python
import math
import random


def minimax(points, i, depth, maximizing, alpha, beta):
  if (depth == 0):
    return points[i]

  if maximizing:
    maxScore = -100000
    for j in range(2):
      score = minimax(points, (2*i)+j, depth-1, False, alpha, beta)
      maxScore = max(score, maxScore)
      alpha = max(alpha, maxScore)
      if (alpha >= beta):
        break
    return maxScore

  else:
    minScore = 100000
    for j in range(2):
      score = minimax(points, (2*i)+j, depth-1, True, alpha, beta)
      minScore = min(score, minScore)
      beta = min(beta, minScore)
      if (alpha >= beta):
        break
    return minScore



id = input("Enter your student ID:")
for i in range(len(id)):
  if (id[i] == '0'):
    id = id[:i] + '8' + id[i+1:]

min_points = int(id[4])
points_to_win = (int(id[-1]) * 10) + int(id[-2])
max_points = math.ceil(points_to_win * 1.5)

random_points = []
for i in range(8):
  random_points.append(random.randrange(min_points, max_points))

print("Generated 8 random points between the minimum and maximum point limits:", random_points)
print("Total points to win:", points_to_win)

achieved_points = minimax(random_points, 0, 3, True, -100000, 100000)

print("Achieved point by applying alpha-beta pruning =", achieved_points)

if (achieved_points >= points_to_win):
  print("The winner is Optimus Prime ")
else:
  print("The winner is Megatron")
```
